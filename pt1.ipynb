## Grammer of Graphics

### Parent layer

Every plot has one.

Defines the plot by linking it to columns in a data frame.

First arugment is the data frame, `data`.

Second argument sets global `aes`thetic/scales by column name reference. These are inherited by child layers.

```python
ggplot(star_wars) # defines the canvas
```

```python
ggplot(star_wars, aes(x = 'height', y = 'mass')) # define aesthetics or scales
```

```python
ggplot(star_wars, aes(x = 'height', y = 'mass', color = 'eye_color'))
```

### Child layers

Two types: geoms and stats.
    
Geoms define one to one mapping from the rows in data to ink on the plot.

Stats perform a summary calculation first, then put that as ink on the plot.

#### Geoms

```python
{
    ggplot(star_wars, aes('height', 'mass'))
    + geom_point()
}
```

#### Stats

```python
{
    ggplot(star_wars, aes('height', 'mass'))
    + geom_point()
    + stat_smooth(method = 'lm')
}
```

```python
{
    ggplot(star_wars, aes('height', 'mass'))
    + geom_point(aes(color = "gender"))
    + stat_smooth(method = 'lm')
}
```

```python
{
    ggplot(star_wars, aes('mass', fill = 'gender'))
    + geom_density() # really a stat 
}
```

```python
{
    ggplot(star_wars, aes('gender', 'mass'))
    + geom_boxplot()
}
```

```python
{
    ggplot(star_wars, aes('gender', 'mass'))
    + geom_point(position = 'jitter')
    + stat_summary(geom = 'crossbar')
}
```

```python
{
    ggplot(star_wars, aes('gender', 'mass'))
    + geom_point(position = position_jitter(width = .2))
    + stat_summary(geom = 'crossbar')
}
```

```python
{
    ggplot(star_wars, aes('gender', 'mass'))
    + geom_point(position = position_jitter(width = .2),
                alpha = .5)
    + stat_summary(geom = 'crossbar')
}
```

```python
{
    ggplot(star_wars, aes('gender', 'mass'))
    + geom_point(position = position_jitter(width = .2),
                alpha = .5)
    + stat_summary(geom = 'crossbar') 
    + scale_y_log10()
}
```
